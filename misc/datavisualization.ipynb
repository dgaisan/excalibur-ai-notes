```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 8, 16, 25]

plt.title("Random plot")
plt.xlabel("Axis X")
plt.ylabel("Axis Y")

plt.plot(x, y, color='green', linestyle="-.", marker="o", label="Green stuff")
plt.plot([0,6,7], [8,9,10], color='red', linestyle="--", marker="*", label="Red stuff")

plt.legend()

plt.show()
```

```python
category = ["A", "B", "C"]
values = [1,5,7]
```

```python
# Histogram
plt.hist([1,1,1,1,2,3,4,5], color="purple", bins=4)
plt.show()
```

```python
# Pie chart
size = [10, 20, 30, 50, 80, 2]
labels = ["a", "b", "c", "d", "e", "f"]
plt.pie(size, labels=labels, autopct="%1.1f%%")
plt.show()
```
