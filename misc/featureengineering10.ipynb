```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = {
    'height': [2.0, 1.0, 1.7, 1.9, 1.75],
    'wight': [80, 60, 75, 100, 55]
}
df = pd.DataFrame(data)
print(f"Original:\n {df}")

min_max_scaler = MinMaxScaler()
df_normilized = pd.DataFrame(min_max_scaler.fit_transform(df), columns=df.columns)
print(f"Normilized:\n {df_normilized}")


standard_scaler = StandardScaler()
df_standard = pd.DataFrame(standard_scaler.fit_transform(df), columns=df.columns)
print(f"Standard:\n {df_standard}")

print(f"mean weight: {df["wight"].mean()}")
print(f"std weight: {df["wight"].std()}")
```

```python
data = {
    "Gender": ["Female", "Female", "Female", "Male", "Female"],
    "Salary": [5000, 4000, 6000, 4000, 7000]
}

df = pd.DataFrame(data)

df_dummies = pd.get_dummies(df, columns=["Gender"])
print(df_dummies)
```

```python
from sklearn.preprocessing import LabelEncoder

df = pd.DataFrame({
    "Colors": ["Red", "Green", "Green","Yellow", "Green"]
})
print(df)
label_encoder = LabelEncoder()

df["Colors"] = label_encoder.fit_transform(df["Colors"])
print(df)
```

```python
df = pd.DataFrame({
    "Colors": ["R", "G", "G","B", "R", "G"]
})

freq = df["Colors"].value_counts()
df["color_freq"] = df["Colors"].map(freq)
print(df)
```

```python
df = pd.DataFrame({
    "Colors": ["R", "G", "G", "B", "R", "G"],
    "Target": [1, 0, 0, 0, 0, 1]
})

target_mean = df.groupby("Colors")["Target"].mean()
df["color_target_mean"] = df["Colors"].map(target_mean)

print(df)
```

```python
from scipy.stats import boxcox

data = {
    "Income": [30000, 35000, 40000, 50000, 100000, 130000]
}

df = pd.DataFrame(data)
df["boxcox"], _ = boxcox(df["Income"])
print(df)

import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 2)
sns.histplot(df["Income"], kde=True)
plt.title("Income normal")

plt.subplot(1, 2, 1)
sns.histplot(df["boxcox"], kde=True)
plt.title("Income BoxCox")
```
